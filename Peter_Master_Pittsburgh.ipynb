```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import date
```

```python
#Reading in the original csv file
csv_path= "Resources/crime_data.csv"
master_df = pd.read_csv(csv_path)
master_df.head()
```

```python
# Clean Pittsburgh Crime DataFrame
# 1.) Split up OFFENSES column into seperate Offenses columns
master_df[['Offense 1','Offense 2','Offense 3','Offense 4','Offense 5']]=master_df['OFFENSES'].str.split(' / ', expand=True,n=4)

# 2.) Split up ARRESTTIME column into Arrest Hour, Arrest Day, Arrest Month, Arrest Year columns

# 3.) Drop unnecessary columns: PK, CCR, ARRESTLOCATION, INCIDENTTRACT, COUNCIL_DISTRICT, PUBLIC_WORKS_DIVISION.
del master_df['PK']
del master_df['CCR']
del master_df['ARRESTLOCATION']
del master_df['INCIDENTTRACT']
del master_df['COUNCIL_DISTRICT']
del master_df['PUBLIC_WORKS_DIVISION']

# 4.) Reorganize & Rename columns:
#     Age, Gender, Race, Arrest Hour, Arrest Day, Arrest Month, Arrest Year, Incident Location, Incident Lat, Incident Lng, Incident Neighborhood, Incident Zone

# 5.) Rename Index as Arrest ID
master_df.index.names = ['Arrest ID']

master_df.head()
```
